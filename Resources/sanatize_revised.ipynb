```python
import pandas as pd
```

```python
cols = pd.read_csv("../Resources/Columns.csv")
data19 = pd.read_csv("../Resources/nhes_19_pfi_v1_0.csv")
data16 = pd.read_csv("../Resources/nhes_16_pfi_v1_0.csv")
data12 = pd.read_csv("../Resources/nhes_12_pfi_v1_0.csv")
```

```python
data19 = data19.append(data16)
data19 = data19.append(data12)
data19.to_csv("compiled.csv")
col_list = []
```

```python
# Load column names to list from CSV
for col, row in cols.iteritems():
     col_list = row.values

col_list
data19.reset_index(drop=True, inplace=True)
data19
```

```python
# Filter columns by new list
new_df = data19.filter(col_list)
len(new_df)
```

```python
# Remove rows with -1 or -9 as a value
values = [-1, -9]
grades = [5]
columns = list(new_df)
print(columns)

for col in columns:
    if (col == "SEGRADES"):
        new_df = new_df[new_df[col].isin(grades) == False]
    else:
        new_df = new_df[new_df[col].isin(values) == False]
new_df
```

```python
# Check for rows with -1 as a value
count = new_df[new_df.columns[1:]].eq(-1).sum().to_frame().T
for col, row in count.iteritems():
    print(col)
    print(row.values)
```
