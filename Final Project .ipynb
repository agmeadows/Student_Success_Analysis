```python
# Trying to make a heat matrix for correlation
corr=school_data_df.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(school_data_df.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(school_data_df.columns, fontsize = 5)
ax.set_yticklabels(school_data_df.columns, fontsize = 5)
plt.show()
```
