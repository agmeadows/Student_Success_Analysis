```python
# import dependencies
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# load csv
mydata_load = "../Resources/nhes_19_pfi_cleaned.csv"
school_data_df = pd.read_csv(mydata_load)
school_data_df = pd.DataFrame(school_data_df)
```

```python
list(school_data_df.columns)
```

```python
#Creating a correlation matrix
school_data_df.corr()
```

```python
# remove allgrade columns and preserve in a separate df
allgrade_df = school_data_df.filter(['ALLGRADEX_2',
 'ALLGRADEX_3',
 'ALLGRADEX_4',
 'ALLGRADEX_5',
 'ALLGRADEX_6',
 'ALLGRADEX_7',
 'ALLGRADEX_8',
 'ALLGRADEX_9',
 'ALLGRADEX_10',
 'ALLGRADEX_11',
 'ALLGRADEX_12',
 'ALLGRADEX_13',
 'ALLGRADEX_14',
 'ALLGRADEX_15'], axis=1)
allgrade_df.head()
```

```python
# remove allgrade columns from school_data_df
school_data_df = school_data_df.drop([ 'ALLGRADEX_2',
 'ALLGRADEX_3',
 'ALLGRADEX_4',
 'ALLGRADEX_5',
 'ALLGRADEX_6',
 'ALLGRADEX_7',
 'ALLGRADEX_8',
 'ALLGRADEX_9',
 'ALLGRADEX_10',
 'ALLGRADEX_11',
 'ALLGRADEX_12',
 'ALLGRADEX_13',
 'ALLGRADEX_14',
 'ALLGRADEX_15'], axis=1)
```

```python
# create features
X = school_data_df.drop(columns = {'SEGRADES_0', 'Unnamed: 0','SEGRADES_1'}, axis=1)

# create target
y = school_data_df['SEGRADES_1']
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
#we analyze how well our random forest model classifies
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
print(f'Accuracy: {acc_score: .2f}')
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_

# We can sort the features by their importance.
sorted(zip(importances, X.columns), reverse=True)
```

```python
# assign importances to variable
Data= sorted(zip(importances, X.columns), reverse=True)

#Creating a dataframe to hold feature importances
feature_df=pd.DataFrame(Data,columns=["value","feature"])

#Trying to identify a number that will improve our outcome
feature_df=feature_df.loc[(feature_df["value"]>.0015)]
feature_list= feature_df["feature"].tolist()

len(feature_list)
```
